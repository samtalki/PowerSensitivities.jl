## Calculate the analytic Jacobian with PowerModels.jl

```julia
function get_pq_bus_idx(data::Dict{String,<:Any})
    n_buses = length(data["bus"])
    return sort([data["bus"][string(i)]["bus_i"] for i in 1:n_buses if data["bus"][string(i)]["bus_type"]==1])
end

function calc_basic_jacobian_matrix_pq(data::Dict{String,<:Any})
    n_buses = length(data["bus"])
    pq_bus_i = get_pq_bus_idx(data) #get the pq bus indeces
    pq_bus_jac_idx_full = [pq_bus_i;pq_bus_i.+(Int(n_buses))] #map the pq bus indeces to jacobian row and column indeces
    jac_full = calc_basic_jacobian_matrix(data)
    jac_truncated = jac_full[pq_bus_jac_idx_full,pq_bus_jac_idx_full] #truncate the jacobian
    return jac_truncated
end

function calc_basic_jacobian_matrix_pq(data::Dict{String,<:Any},jac_precalc_full::Matrix)
    n_buses = length(data["bus"])
    pq_bus_i = get_pq_bus_idx(data) #get the pq bus indeces
    pq_bus_jac_idx_full = [pq_bus_i;pq_bus_i.+(Int(n_buses))] #map the pq bus indeces to jacobian row and column indeces
    jac_truncated = jac_precalc_full[pq_bus_jac_idx_full,pq_bus_jac_idx_full] #truncate the jacobian
    return jac_truncated
end
```

```julia
pq_buses = get_pq_bus_idx(pm_data_dict)
jac_pq = calc_basic_jacobian_matrix_pq(pm_data_dict)
```

```julia
n_PQ_buses = length(pq_buses)
jac_vmag = jac_pq[:,n_PQ_buses+1:end] #Get Vmag sensitivity block matrices
print("Voltage magnitude sensitivity block matrices",size(jac_vmag),jac_vmag)
```
